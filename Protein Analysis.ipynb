In addition to the typical NormalDistribution, two other distributions had to be defined. The first is <a href=http://en.wikipedia.org/wiki/Log-normal_distribution>LogNormalDistribution</a>, which is just a typical lognormal distribution. The second is ExtremeMinimumDistribution, which is the <a href=http://en.wikipedia.org/wiki/Generalized_extreme_value_distribution>left-skewed version of the general extreme distribution</a>. These are both going to be extensions of the abstract Distribution class from <a href=https://github.com/jmschrei/yahmm>YAHMM</a>. 

```python
class ExtremeMinimumDistribution( Distribution ):
    '''
    A Gumbel Type I distribution, or Extreme Minimum.
    '''

    def __init__( self, mu, beta ):
        '''
        Mu is the location parameter, and beta is the scale parameter for the
        distribution.
        '''

        self.parameters = [ float(mu), float(beta) ]

    def log_probability( self, symbol ):
        '''
        Calculate the log probability of a symbol under this distribution.
        '''

        scale = ( symbol - self.parameters[0] ) / self.parameters[1]
        return -np.log( self.parameters[1] ) + scale - math.exp( scale )

class LogNormalDistribution( Distribution ):
    '''
    A log normal distribution.
    '''

    def __init__( self, mu, sigma ):
        self.parameters = [ float( mu ), float( sigma ) ]

    def log_probability( self, symbol ):
        '''
        Return the log probability of this symbol under the model.
        '''

        mu, sigma = self.parameters
        return -np.log( symbol * sigma * np.sqrt( 2 * np.pi ) ) - 0.5 * ( ( np.log( symbol ) - mu ) / sigma  ) ** 2
```

Now lets build the profile. There are 29 states, including open channel being represented in the first state. This is useful if we want to normalize the data. I choose not to here. The profile is represented by two spreadsheets, one with means and one with standard deviations. Columns represent profile positions and rows represent events. Using pandas it's extremely simple to grab this data.

```python
profile_means = pd.read_csv( 'C:\\Users\\jmschrei\\Desktop\\Protein Automation Data\\s2gt_mean_revised.csv' )
profile_stds = pd.read_csv( 'C:\\Users\\jmschrei\\Desktop\\Protein Automation Data\\s2gt_std_revised.csv' )

dists_means = [ NormalDistribution( profile_means[col].mean(), profile_means[col].std()*3 ) for col in profile_means ][1:]
dists_stds = [ LogNormalDistribution( np.log( profile_stds[col] ).mean(), np.log( profile_stds[col] ).std()*2 ) for col in profile_stds ][1:]
```

We discard the first distribution, as it represents open channel, and we don't want that.

Means are represented using normal distributions, and standard deviations are represented using log normal distributions. One of the reasons log normal distributions are a good choice for standard deviations is that they are both bounded by 0 as a minimum, and that it becomes exponentially less likely the closer it gets to 0. Usually they clusters approximately around 1, but are more likely to be 1.1 then 0.9. The standard deviation of both distributions are scaled by 3 and 2 respectively. This is an arbitrary decision on my part, because otherwise the data does not align well to the model. This will have to be corrected eventually.

Now lets build the actual profile which has two independent components-- mean and standard deviation.

```python
profile = [ MultivariateDistribution([ mean, std ]) for mean, std in it.izip( dists_means, dists_stds ) ]
profile[14] = MultivariateDistribution([ ExtremeMinimumDistribution( 14, 2 ), LogNormalDistribution( np.log(2), np.log(0.5) ) ])
```

The 15th position in the profile needs to be different. We'll see why in a trace later.

Now lets define a simple insert distribution as a uniform distribution across reasonable values. The modular subunit which I chose was the generic Phi29 one. You may choose to write your own, depending on what you think the phenomena of ClpX are. 

```python
insert = MultivariateDistribution( [ UniformDistribution( 0, 40 ), UniformDistribution( 0, 10 ) ] )
model = ModularProfileModel( Phi29GlobalAlignmentModule, profile, "ClpXProfile-{}".format( len(profile) ), insert )
```

Now we have the meat of the function, once we've defined the model.

For each file in a list of files, parse the file to get the events. Then for each event, filter using a 1st order 2000Hz low pass bessel filter, segment it using the frequentist interpretation with an appropriate threshold on the score function. Then we want to plot the event segmented, the event colored by the HMM, and the corresponding colors for each state.

The reason we calculate `_, hidden_states = model.viterbi....` is because the generic plot function assumes you want to run a sequence of the means through a 1D model. This is not something easily generalizable, so if you want to create your own path of hidden states you can pass them in under the `hidden_states` argument. This is useful because we're defining a distribution across multiple dimensions.

```python
filenames = ["C:\\Users\\jmschrei\\Desktop\\Protein Automation Data\\13n25010-s05.abf"] 

for file in it.imap( File, filenames ): 
    print file.filename
    file.parse( lambda_event_parser( threshold=40 ) )

    for event in file.events:
        event.filter()
        event.parse( SpeedyStatSplit( false_positive_rate=1e-3, min_width=5, cutoff_freq=2000 ) )

        _, hidden_states = model.viterbi( np.array( [ [ seg.mean, seg.std ] for seg in event.segments ] ) ) 

        plt.figure( figsize=(20, 8))
        plt.subplot( 311 )
        event.plot( color='cycle' )

        plt.subplot( 312 )
        event.plot( color='hmm', hmm=model, hidden_states=hidden_states )

        ax = plt.subplot( 313 )
        plt.imshow( [ np.arange( 0., len(profile) ) / len(profile) ], interpolation='nearest', cmap="Set1" )
        plt.grid( False )

        means = [ d.parameters[0][0].parameters[0] for d in profile ]
        
        for i, mean in enumerate( means ):
            plt.text( i-0.2, 0.1, str( round(mean, 1) ), fontsize=16 )

        #print "\n".join( state.name for idx, state in event.apply_hmm( model, algorithm='viterbi')[1] )

        plt.show()
    file.close()
```
